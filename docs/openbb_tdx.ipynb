```python
from openbb import obb

if 'info' in obb.reference:
    obj = obb.reference["info"]["extensions"]["openbb_provider_extension"]
    print([item for item in obj if "tdx" in item])

symbols = "601919,600519,000001,00300"
symbols_a = "601919,600519,000001,600325,600028,600036,430047"
symbol_a = "000001"
symbol_hk = "01658"
provider = "tdxquant"
start_date="2025-09-08"
end_date="2025-12-31"

```

## 1. 股票相关模型 (Equity Models)
### 1.1 Equity Historical Price (股票历史价格)

```python
obb.equity.price.historical(symbol=symbol_a, provider=provider, use_cache=False, start_date=start_date, end_date=end_date).to_dataframe().tail()
```

### 1.2 Equity Quote (股票实时报价)

```python
obb.equity.price.quote(symbols, provider=provider).to_dataframe().head()
```

### 1.3 Historical Dividends (历史分红)

```python
obb.equity.fundamental.dividends(symbol=symbol_a, provider=provider, use_cache=False).to_dataframe().head()
```

### 1.4 Equity Profile (股权资料)

```python
obb.equity.profile(symbol="00005", provider=provider, use_cache=True).to_dataframe().tail()
```

### 1.4 Equity Search (股票搜索)

```python
# A-shares symbols
obb.equity.search(query="", market="5", provider=provider).to_dataframe()
```
